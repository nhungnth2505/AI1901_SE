```python
import requests
import sqlite3

# 1. Crawl (RSS feed)
url = "https://tuoitre.vn/rss/tin-moi-nhat.rss"
response = requests.get(url)

# 2. Conversion
content = response.content.decode("utf-8", errors="ignore")

print(" Một phần dữ liệu RSS")
print(content[:500])

# 3. Store vào Database SQLite
conn = sqlite3.connect("docs.db")
cur = conn.cursor()

cur.execute("CREATE TABLE IF NOT EXISTS documents (url TEXT, content TEXT)")
cur.execute("INSERT INTO documents VALUES (?, ?)", (url, content))
conn.commit()

print("\n Đã lưu dữ liệu vào DB ")

# Truy vấn thử
cur.execute("SELECT url, substr(content,1,200) FROM documents LIMIT 1")
row = cur.fetchone()
print("\n Truy xuất từ DB")
print("URL:", row[0])
print("Content preview:", row[1])
```
